```python
#lista de parametros para API

import requests
import pandas as pd
import numpy as np
import time

response = requests.get('https://servicodados.ibge.gov.br/api/v3/agregados?nivel=N6')

if response.status_code == 200:
    data = response.json()

    resultado = pd.DataFrame({'nivel_0':[],'tabela_id':[]})

    data_list = []

    for item in data:
        id = item['id']
        pesquisa = item['nome']
        for agregado in item['agregados']:
            tabela = agregado['id']
            nome_tabela = agregado['nome']
            data_list.append([id, pesquisa, tabela, nome_tabela])
            df = pd.DataFrame(data_list, columns=['id', 'pesquisa', 'tabela', 'nome_tabela'])
    lista = df['tabela']

    #lista = pd.DataFrame({'tabela':['6784','1621','2093']})
    print(lista)
    for item in lista:
        tab = item
        enderecoT = 'https://servicodados.ibge.gov.br/api/v3/agregados/'+tab+'/metadados'
        print(enderecoT)
        responseT = requests.get(enderecoT)
        dataT = responseT.json()

        df_var = pd.json_normalize(dataT,
            record_path=['variaveis'],
            meta=["id"],
            meta_prefix='tabela_',
            record_prefix='var_'
                    )

        print(df_var)
else: 
    print(f"Request failed with status code {response.status_code}")
```
