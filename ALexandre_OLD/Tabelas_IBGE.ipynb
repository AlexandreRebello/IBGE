```python
import requests
import pandas as pd

response = requests.get(
    'https://servicodados.ibge.gov.br/api/v3/agregados/1621/metadados')

if response.status_code == 200:
    data = response.json()

    data_list = []

    for item in data:
        id = data['id']
        pesquisa = data['pesquisa']
        tabela = data['nome']
        periodicidade = data["periodicidade"]["frequencia"]
        per_inicio = data["periodicidade"]["inicio"]
        per_fim = data["periodicidade"]["fim"]
        nivel_territorial= data["nivelTerritorial"]["Administrativo"]
        nivel_territorial_especial= data["nivelTerritorial"]["Especial"]
        nivel_territorial_IBGE= data["nivelTerritorial"]["IBGE"]
        for classificacao in data['classificacoes']:
            class_id = classificacao['id']
            class_nome = classificacao['nome']
            for categoria in classificacao['categorias']:
                cat_id = categoria['id']
                cat_nome = categoria['nome']
                cat_nivel = categoria['nivel']
                data_list.append([id,pesquisa,tabela,periodicidade,per_inicio, per_fim, nivel_territorial, nivel_territorial_especial, nivel_territorial_IBGE,class_id,class_nome,cat_id,cat_nome,cat_nivel])

    df = pd.DataFrame(data_list, columns=[
                      'id_tabela','pesquisa', 'nome_tabela','periodicidade','inicio','fim','nivel_territorial', 'nivel_territorial_especial', 'nivel_territorial_IBGE','id_classificacao', 'nome_classificacao', 'id_categoria', 'nome_categoria', 'nivel_cat'])

    print(df)
else:
    print(f"Request failed with status code {response.status_code}")
```
