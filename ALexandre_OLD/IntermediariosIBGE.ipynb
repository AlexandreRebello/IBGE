```python
import requests
import pandas as pd

divisoes= requests.get("https://servicodados.ibge.gov.br/api/v2/cnae/divisoes").json()
divisoes
secao_id = [ divisoes["id"] for divisoes in divisoes ]
secao_descr = [ divisoes["descricao"] for divisoes in divisoes ]
divisao_id = [ divisoes["secao"]["id"] for divisoes in divisoes ]
divisao_descr =  [ secao["secao"]["descricao"] for secao in divisoes ]
divisoes_base = pd.DataFrame.from_dict({'Div_ID':divisao_id,'Div_Descricao':divisao_descr,'Secao_ID':secao_id,'Secao_descricao':secao_descr})
divisoes_base
```
